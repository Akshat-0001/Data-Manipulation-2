```python
# 1.1 Create a series using list
import pandas as pd

a1 = [1, 3, 5, 7, 9, 2, 4, 6, 8]
a2 = pd.Series(a1) #holds data of any type.

print(a2)
```

```python
#1.2 A series has been created with data along with it's Index
import pandas as pd

a1 = [1, 3, 5, 7, 9, 2, 4, 6, 8]
a2 = ['a','b','c','d','e','f','g','h','i']
a3 = pd.Series(a1, a2) #a2 is considered as index
#a3 = pd.Series(a2, a1) #a1 is considered as index
print(a3)
```

```python
print ("a3['b'] -->",a3['b'])
print ("a3['i'] -->",a3['i'])
```

```python
#1.3 Creating a series using dictionary
import pandas as pd

d1 = {'Guavas':5, 'Pineapples':7, 'Grapes':13, 'Strawberries':8}
d2 = pd.Series(d1)

print (d2)
print (type(d2))
```

```python
#1.4 Creating a series using nested list
import pandas as pd

a1 = [[1,3,5],[2,4,6]]
a2 = pd.Series(a1)

print (a2)
```

```python
#2.1 Creating a data frame using dictionary
import pandas as pd

d1 = {'Age':[18,21,27,38],'Name':['Akshat','Kira','Itadori','Gojo']}
d2 = pd.DataFrame(d1)

d2
```

```python
#2.2 Creating a data frame using nested list
import pandas as pd

d1 = [[4,1900],[3,1600],[2,1100],[1,850]]
d2 = pd.DataFrame(d1, columns = ['Classrooms','Area'])

d2
```

```python
#2.3 Assigning indexes within a data frame
import pandas as pd

d1 = {'Name':['Shinchan','Sasaki','Haise','Sushi'],'Marks':[73,69,96,59]}
d2 = pd.DataFrame(d1,index = ['Rank 2','Rank 3','Rank 1','Rank 4'])

d2
```

```python
#2.5 Creating random data frame
import numpy as np
import pandas as pd
import random as r
import string as s

df = pd.DataFrame({'A': [r.randint(1, 9)     for x in range(5)],
                   'B': [r.randint(1, 9)*10  for x in range(5)],
                   'C': [r.randint(1, 9)*100 for x in range(5)],
                   'D': [ "".join(r.sample(s.ascii_letters,5)) for x in range(5)],
                   'E': 'Ohayo',
                   'F': 1})

df
```

```python
#2.6 Describe the data frame
df.describe() 
```
